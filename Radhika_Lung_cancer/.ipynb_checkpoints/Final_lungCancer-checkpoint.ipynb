# **LABEL ENCODING CATEGORICAL VARIABLES**

```python
df.GENDER = df.GENDER.map({"M":1,"F":0})
df.LUNG_CANCER = df.LUNG_CANCER.map({"YES":1,"NO":0})
```

```python
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
#dependent variable
x = df.iloc[:,0:-1]
x
```

```python
#independent_variable
y = df['LUNG_CANCER']
y
```

```python
#Splitting the Dataset: Training and Testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Now check the shape of split data
print(x_train.shape, x_test.shape) # dependent Data
print(y_train.shape, y_test.shape) # independent data
```

# **Data preprocessing is done and also done split of data into train test Now we will Train our model**

# **IMPORT THE LOGISTICS REGRESSION MODEL**

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(x_train, y_train)
```

```python
 # Now check the training score of model
log_model.score(x_train, y_train)
```

```python
lr_y_pred = log_model.predict(x_test)
lr_y_pred
```

```python
# Model Evaluation on the basis of confusion matrix, accuracy score, classification report
# import of evalation matrix

from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

def model_evaluation(y_actual, y_prediction):
    print('Accuracy_Score = ',accuracy_score(y_actual, y_prediction))
    print('Confusion Matrix \n',confusion_matrix(y_actual, y_prediction))
    sns.heatmap(confusion_matrix(y_actual, y_prediction), annot = True, fmt = '.2f')
    print('Classification_Report \n',classification_report(y_actual, y_prediction))
```

```python
model_evaluation(y_test, lr_y_pred)
```

```python
test_acc = [98.38]
model_name = ['Logistic_Regression']
```

# Improve KNN model by changing values of K
# Find best K value

```python
er = []
test_acc = []
model_acc = []

for i in range(3,62,2): # 
    knn = KNeighborsClassifier(n_neighbors= i,weights='distance' )
    knn.fit(x_train, y_train)
    y_pred = knn.predict(x_test)
    er.append(np.mean(y_test!=y_pred))
    test_acc.append(accuracy_score(y_test, y_pred))
    model_acc.append(knn.score(x_train,y_train))
```

```python
# to check the lowest error rate
plt.figure(figsize = (10,8))
plt.plot(range(3,62,2), er, marker = 'o')
plt.title('Error V/s K_value')
plt.xlabel('K Value')
plt.ylabel('Error')
plt.xticks(range(3,62,2))
plt.yticks(er)
plt.show()
```

```python
# to check the highest Accuracy
plt.figure(figsize = (10,8))
plt.plot(range(3,62,2), test_acc, marker = 'o')
plt.title('Accuracy V/s K_value')
plt.xlabel('K Value')
plt.ylabel('Accuracy')
plt.xticks(range(3,62,2))
plt.yticks(test_acc)
plt.show()
```
